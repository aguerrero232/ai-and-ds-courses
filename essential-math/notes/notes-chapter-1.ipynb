## Derivatives

A derivative tells the slope of a function, and it is useful to
measure the rate of change at any point in a function.

```python
# a derivative function in python
def derivative_x(f, x, step_size):
    return (f(x + step_size) - f(x)) / ((x + step_size) - x)

def my_function(x):
    return x**2

slope_at_2 = derivative_x(my_function, 2, .00001)
slope_at_2 # 4.000010000000827
```

Calculate the derivative using sympy

```python
x = symbols("x")

f = x**2

dx_f = diff(f)
dx_f # 2*x
```

## Partial Derivatives

Derivatives of functions that have multiple input
variables. Rather than finding the slope on a one-dimensional function,
we have slopes with respect to multiple variables in several directions. For each given
variable derivative, we assume the other variables are held constant. 

```python
x,y = symbols("x y")

# the function
f = 2*x**3 + 3*y**3

# calc the derivatives
dx_f = diff(f, x)
dy_f = diff(f, y)

print(f"derivative of f with respect to x: {dx_f}")
print(f"derivative of f with respect to y: {dy_f}")

plot3d(f)
```

## Using limits to calculate derivatives

```python
# Using limits to calculate a slope
x, s = symbols("x s")

f = x**2

slope_f = (f.subs(x, x + s) - f) / ((x+s) - x)
slope_2 = slope_f.subs(x, 2)

# calc slope at x = 2
result = limit(slope_2, s, 0)
result
```

```python
# Using limits to calculate a derivative
x, s = symbols("x s")

f = x**2

slope_f = (f.subs(x, x + s) - f) / ((x+s) - x)

# calc the derivative of the function
# infinitely approach step size +s+ to 0
result = limit(slope_f, s, 0)
result # the derivative of the function x^2 is 2x
```

## Chain Rule


```python
z = (x**2 + 1)**3 -2

dz_dx = diff(z, x)
dz_dx # the derivative of z with respect to x is 6x(x^2 + 1)^2 where z = (x^2 + 1)^3 - 2
```

Calculating the derivative dz/dx with and without the chain rule

```python
x, y = symbols("x y")

# derivative of the first function
_y = x**2 + 1
dy_dx = diff(_y)

# derivative of the second function
z = y**3 - 2
dz_dy = diff(z)

# calc the derivative with and without chain rule, sub y function
dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
dz_dy_no_chain = diff(z.subs(y, _y))

# the chain rule is proved true if both are equal
print(f"chain rule: {dz_dx_chain}")
print(f"no chain rule: {dz_dy_no_chain}")
```

## Integrals

```python
# an integral approximation function in python
def approximate_integral(a, b, n, f):
    delta_x = (b - a) / n
    total_sum = 0
    for i in range(1, n + 1):
        midpoint = 0.5 * (2 * a + delta_x * (2 * i - 1))
        total_sum += f(midpoint)
    return total_sum * delta_x

def my_function(x):
    return x**2 + 1
```

```python
# num rectangles = 5
approximate_integral(a=0, b=1, n=5, f=my_function)
```

```python
# num rectangles = 1000
approximate_integral(a=0, b=1, n=1000, f=my_function)
```

```python
# num rectangles = 1000000
approximate_integral(a=0, b=1, n=1000000, f=my_function)
```

### Calculating Integrals with Limits

The main idea follows much of what we did earlier: pack rectangles under a curve
and make them infinitely smaller until we approach the exact area.

```python
x, i, n = symbols('x i n')

f = x**2 + 1
lower, upper = 0, 1

# calculate width and each rectangle height at index 'i'
delta_x = ((upper - lower) / n)
x_i = (lower + delta_x * i)
fx_i = f.subs(x, x_i)

# iterate over n rectangles and sum the area
n_rectangles = Sum(delta_x * fx_i, (i, 1, n)).doit()

# calculate the area by approaching the number of rectangles n to infinity
area = limit(n_rectangles, n, oo)
area
```
